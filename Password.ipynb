```python
condition1 = "#" in password
condition2 = len(password) >= 8
condition3 = len(password) <= 20
```

```python
if condition1:
  print("Password is valid") #this will print even if the length is wrong, since the only condition for this is condition1
else:
    print("Password needs a #")
    if condition2:
      print("Password is valid") #this will print even if condition3 is not met
    else:
        print("Password needs 8 or more characters or ") #this will not print unless the # is also missing, even if we have under 8 chars
        if condition3:
          print("Password is valid") #this will still print if and only if condition 1 is NOT met
        else:
            print("Password needs 20 or less characters") #this will only print if there's also no # in the password and the password is less than 8 characters


#remember, every time you indent an if statement further than a previous if statement:
  # you are telling the computer to only check that condition if the parent statement was triggered
```
