```python
# Use the input function to request so items from the user
price_of_tv = float(input("price of tv"))
price_of_playstion5 = float(input("price of playstion5"))
price_of_blackops7 = float(input("price of blackops7"))
```

```python
# Extracting datatypes through SQA prints
print(type(price_of_tv))
print(type(price_of_playstion5))
print(type(price_of_blackops7))
```

```python
# Making a variable for the total price and adding and subtracting it
total_price = price_of_tv + price_of_playstion5 + price_of_blackops7
print(total_price)
```

```python
# Adding new variables to the total price
price_of_gamimg_chair = float(input("price of gaming chair"))
price_of_gaming_desk = float(input("price of gaming desk"))
total_price = total_price + price_of_gamimg_chair + price_of_gaming_desk
print(total_price)
```

```python
# Coverting the total price to a integer and printing out the total price and subtracting the float
total_price = int(total_price)
print(total_price)
```
